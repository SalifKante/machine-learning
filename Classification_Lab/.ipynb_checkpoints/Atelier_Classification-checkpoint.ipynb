# Variables X, y et la répartition des classes

```python
X = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y
```

```python
print('Pourcentage de 0 est {0:.3f}'.format(100*sum(y==0)/len(y)))
print('Pourcentage de 1 est {0:.3f}'.format(100*sum(y==1)/len(y)))
print('Pourcentage de 2 est {0:.3f}'.format(100*sum(y==2)/len(y)))
```

```python
plt.hist(y)
plt.title('Répartition des classes')
```

# Train and test sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
X_test.shape
```

# Classifieurs

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
def classifier(X_train,X_test,y_train,y_test):
    
    DT = DecisionTreeClassifier(random_state=0)
    DT.fit(X_train,y_train)
    y_pred_DT = DT.predict(X_test)
    print('Confusion matrix DT {}'.format(confusion_matrix(y_test,y_pred_DT)))
    print('Accuracy score DT {}'.format(accuracy_score(y_test,y_pred_DT)))
    
    print('\n')
    
    RF = RandomForestClassifier(n_estimators = 40, random_state=0)
    RF.fit(X_train,y_train)
    y_pred_RF = RF.predict(X_test)
    print('Confusion matrix RF {}'.format(confusion_matrix(y_test,y_pred_RF)))
    print('Accuracy score RF {}'.format(accuracy_score(y_test,y_pred_RF)))
```

```python
classifier(X_train,X_test,y_train,y_test)
```

# Feature importance with Random Forest

```python
clf = RandomForestClassifier(n_estimators=500, random_state=0)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
name_col = dataset.columns[0:5]
print(name_col)
```

```python
importances = clf.feature_importances_
```

```python
sorted_idx = np.argsort(importances[::-1])
```

```python
print(importances)
```

```python
features = name_col
print(features[sorted_idx])
```

```python
padding=np.arange(X_train.size/len(X_train))+0.5
plt.barh(padding,importances[sorted_idx],align='center')
plt.yticks(padding,features[sorted_idx])
plt.xlabel("Relative importance")
plt.title("Variable importance")
plt.show()
```

# Visualize Decision Boundaries

```python
from adspy_shared_utilities import plot_class_regions_for_classifier_subplot

#X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 0)
fig, subaxes = plt.subplots(6, 1, figsize=(6, 32))

pair_list = [[0,1], [0,2], [0,3], [1,2], [1,3], [2,3]]
tree_max_depth = 4

for pair, axis in zip(pair_list, subaxes):
    X = X_train[:, pair]
    y = y_train
    
    clf = DecisionTreeClassifier(max_depth=tree_max_depth).fit(X, y)
    title = 'Decision Tree, max_depth = {:d}'.format(tree_max_depth)
    plot_class_regions_for_classifier_subplot(clf, X, y, None,
                                             None, title, axis,
                                             iris.target_names)
    
    axis.set_xlabel(iris.feature_names[pair[0]])
    axis.set_ylabel(iris.feature_names[pair[1]])
    
plt.tight_layout()
plt.show()
```
